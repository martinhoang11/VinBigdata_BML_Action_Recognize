```python
def convert_data_to_df(data, frames, key_points):
    result = []
    label = []
    for fname in data['file_name'].unique():
        temp = []
        for index, row in data[data['file_name'] == fname].iterrows():
            for i in range(2, key_points+2):
                xy = data.iloc[index][i].replace("['", '').replace("']", '').split("', '")
                xy = [float(j) for j in xy] 
                x = xy[0]
                y = xy[1]
                temp.append(x)
                temp.append(y)
            if len(temp) == frames*key_points*2:
                result.append(temp)
                label.append(data['label'].iloc[index])
                temp = []
                
    df_result = pd.DataFrame(result)
    df_result['label'] = label
    return df_result
```
